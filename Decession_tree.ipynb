It has a hierarchical, tree structure, which consists of a root node, branches, internal nodes and leaf nodes.

Types of Decision Trees :

Decision Tree : ( DTs - Classification )

A decision tree classifier is a non - parametric supervised learning algorithm that uses a tree structure to classify instances based on their feature values. It is a rule-based approach that uses the values in each feature to split the dataset.

A decision tree has a hierarchical tree structure with a root node, branches, internal nodes, and leaf nodes. The nodes in a decision tree represent instances, the branches represent the outcomes of the test, and the leaf nodes represent the class label.

Decision trees are used for classification and regression tasks. They are most preferred for solving classification problems.

```python
from sklearn.tree import DecisionTreeClassifier
DT_C =DecisionTreeClassifier()
```

```python
DT_C.fit(x_train,y_train)
```

**Evaluation Metrics**
Metrics are the measures that evaluate the model we built in ML. Based on the metrics score and prediction accuracy, model is said to be well built or not.

The point to be noted here is every metrics can not be applied for every other problem. Evaluation metrics help in fine tuning to optimize the performance of the model, which results in good



**Metrics for Classification**

Classification metrics are as follows :


Confusion matrix

Accuracy

Recall

Precision

F1 score

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
```

```python
y_predict = DT_C.predict(x_test)
y_predict
```

```python
# calculate the confusion matrix

confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
# Calculate the accuracy

accuracy = accuracy_score(y_test,y_predict)
# accuracy = np.sum(confusion_matrix.diagonal()) / np.sum(confusion_matrix)
accuracy
```

```python
# Calculate the precision

# precision = precision_score(y_test,y_predict)
precision = confusion_matrix[1, 1] / (confusion_matrix[1, 1] + confusion_matrix[0, 1])
precision
```

```python
# Calculate the recall

# recall = recall_score(y_test,y_predict)
recall = confusion_matrix[1, 1] / (confusion_matrix[1, 1] + confusion_matrix[1, 0])
recall
```

```python
# Calculate the F1 score

#f1_score = f1_score(y_test,y_predict)
f1_score = 2 * precision * recall / (precision + recall)
f1_score
```
